# Source activation

Simple plot, just a sigmoid with various values of gradient parameters.

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib

# Some Parameters
figsize = (5,2.5)
dpi     = 200
Npts    = 100
xlim    = (-5,5)
A       = [1,2,10,1e2]


# Create the data
s  = lambda x,a: 1/(1+np.exp(-a*x))
x  = np.linspace(*xlim,Npts)
ys = [s(x,a) for a in A]

# Create the plot
fig = plt.figure(figsize=figsize,dpi=dpi)
ax  = fig.add_subplot(111)

# Plot the stuff
for y in ys:
    ax.plot(x,y,c='k',lw=.5)

plt.show()

tikzplotlib.save('test.tex',fig)
# Customize the plot
```
