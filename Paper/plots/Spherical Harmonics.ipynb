# Spherical Harmonic Plots

Here we create the plots of the spherical harmonics used in section 3 of the paper in order to illustrate the different physical categories of decay.


## Underdamped
The first case we will examine is the underdamped case where we can supperpose an outgoing and an ingoing wave from the origin. Here we will plot the ougoing waves. The equation of the two classes of solutions based on their singularities are as follows:

$$P_1(r,t) = \frac{e^{-k^2 c^2\, t/2\omega_0}}{r} cos(kr - \omega(k) t)$$

and,

$$P_2(r,t) = \frac{e^{-k^2 c^2\, t/2\omega_0}}{r} sin(kr - \omega(k) t)$$

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as c
import os
# os.environ['PATH'] = os.environ['PATH'] + ':/Library/TeX/texbin/latex'

# Define some useful constants
w0 = 2
c  = 1
k0 = 2 * w0/c


# Define the functions
w  = lambda k     : k*c * np.sqrt(1 - k**2 * c**2/(4 * w0**2))
w_ = lambda k     : k*c
P1 = lambda r,t,k : np.exp(-k**2 * c**2 * t/ (2* w0))/r * np.cos(k*r - w(k)*t)
P1_= lambda r,t,k : 1/r * np.cos(k*r - w_(k)*t)
P2 = lambda r,t,k : np.exp(-k**2 * c**2 * t/ (2* w0))/r * np.cos(k*r - w(k)*t)
P2_= lambda r,t,k : 1/r * np.sin(k*r - w_(k)*t)
```

```python
# Now we plot

# Ranges
Npts = 1000
t = np.linspace(1e-1,2e+1,Npts)
T = 1
r = np.linspace(1e-10,2e+1,Npts)
R = np.mean(r)
k = 2

# Plottables

p1 = P1(r,T,k)
p1_= P1_(r,T,k)

# Now the figure
plt.rc('font', family='monospace')
fig = plt.figure(figsize=(10,4),dpi=200)
fig.suptitle('Underdamped Case VS Undamped Case')

# Axis 1
ax1 = fig.add_subplot(111)
ax1.set_ylim(-0.5,1.25)
ax1.set_xlim(min(r)-0.2,max(r))
ax1.tick_params(direction='in', bottom=False, left=False, top=False, right=False)

ax1.plot(r,p1 ,lw=1,c='k',label='Underdamped')
ax1.plot(r,p1_,lw=1,c='grey',ls=":",label='Undamped')
ax1.text(10,-0.4,'$\omega_0 = 2$\t\t $c=1$\t\t $k = 2$\t $t=1$',fontsize=12,horizontalalignment='center',fontweight='bold')


ax1.legend(frameon=False,fontsize=12);
```

## Frequency plot

We will also compare the frequencies between the undamped and underdamped regimes in order to show how the dispersion relation is affected as part of the magical thing that we are doing.

Specifically in the underdamped case the dispersion relation is given by:

$$\omega(k) = k c \sqrt{1-\left(\frac{kc}{2\omega_0}\right)^2}$$

Now we will attempt to plot it.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as c

# Define some useful constants
w0 = 2
c  = 1
k0 = 2 * w0/c


# Define the functions
w  = lambda k     : k*c * np.sqrt(1 - k**2 * c**2/(4 * w0**2))
w_ = lambda k     : k*c
```

```python
# Now we plot

# Ranges
Npts = 1000
k = np.linspace(-k0,k0,Npts)

# Data

W = w(k)
W_= w_(k)

# Figure
# plt.rcParams.update({
#     "text.usetex": True
# #     "font.family": "serif",
# #     "font.serif": ["Palatino"],
# })
fig = plt.figure(figsize=(10,6),dpi=200)
# fig.suptitle('$\omega(k)$ in Underdamped and Undamped regimes')

# Axis 1
ax1 = fig.add_subplot(111)
ax1.set_title(r'$\omega(k)$ in Underdamped and Undamped regimes')
# ax1.set_ylim(-0.5,1.25)
# ax1.set_xlim(min(r)-0.2,max(r))
# Grid
ax1.tick_params(direction='in', bottom=True, left=True, top=True, right=True)
ax1.grid(linestyle='--',linewidth=0.4)

# Ticks
xticks = [r'$-k_0$',r'$-\frac{k_0}{2}$',r'$0$',r'$\frac{k_0}{2}$',r'$k_0$']
ax1.set_xticks(np.linspace(-k0,k0,5))
ax1.set_xticklabels(xticks)
ax1.set_xlabel(r'Wavenumber $k\ /\ k_0$')
yticks = [r'$-k_0c$',r'$-\frac{k_0 c}{2}$',r'$0$',r'$\frac{k_0 c}{2}$',r'$k_0 c$']
ax1.set_yticks(np.linspace(-k0,k0,5))
ax1.set_yticklabels(yticks)
ax1.set_ylabel(r'Frequency $\omega(k)\ /\ k_0 c$')

# Plots
ax1.plot(k,W ,lw=1,c='k',label='Underdamped')
ax1.plot(k,W_,lw=1,c='grey',ls=":",label='Undamped')
# ax1.text(10,-0.4,'$\omega_0 = 2$\t\t $c=1$\t\t $k = 2$\t $t=1$',fontsize=12,horizontalalignment='center',fontweight='bold')

ax1.legend(frameon=False,fontsize=12);
```

## Time constant

Here we calculate the overdamped time constant for the system to first order in time as a function of the wavenumber k. The analytical expression for this is as follows:

$$\tau(k)= \left[kc \left(\frac{k}{k_0} + \sqrt{\frac{k}{k_0}-1} \right)\right]^{-1}$$

hence we can plot below.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as c

# Define some useful constants
w0 = 2
c  = 1
k0 = 2 * w0/c


# Define the functions
tau = lambda k: (k*c * (k/k0 + np.sqrt(k/k0-1)))**(-1)
```

```python
# Now we plot

# Ranges
Npts = 1000
k = np.linspace(k0,3*k0,Npts)

# Data
T = tau(k)

# Figure
fig = plt.figure(figsize=(10,6),dpi=200)

# Axis 1
ax1 = fig.add_subplot(111)
ax1.set_title(r'Time constant $\tau(k)$ as a function of wavenumber')
ax1.set_ylim(0,1.1/(k0*c))
ax1.set_xlim(0.8*k0,3*k0)

# Grid
ax1.tick_params(direction='in', bottom=True, left=True, top=True, right=True)
ax1.grid(linestyle='--',linewidth=0.4)

# Ticks
xticks = [r'$k_0$',r'$2k_0$',r'$3k_0$']
ax1.set_xticks(np.linspace(k0,3*k0,3))
ax1.set_xticklabels(xticks)
ax1.set_xlabel(r'Wavenumber $k\ /\ k_0$')
yticks = [r'$0$',r'$\frac{1}{2k_0c}$',r'$\frac{1}{k_0 c}$']
ax1.set_yticks(np.linspace(0,1/(k0*c),3))
ax1.set_yticklabels(yticks)
ax1.set_ylabel(r'Decay Time constant $\tau(k)\ /\ \frac{1}{k_0c}$')

# Plots
ax1.plot(k,T,lw=1,c='k')
ax1.plot([k0,k0],[0,1/(k0*c)],c='darkgrey',ls='--',lw=2)
ax1.plot([0,k0],[1/(k0*c)]*2,c='darkgrey',ls='--',lw=2)
ax1.text(2.5*k0,0.75/(k0*c),r'$\tau(k) \leq \frac{1}{k_0 c}$',fontsize=15,horizontalalignment='center',fontweight='bold');

# ax1.legend(frameon=False,fontsize=12);
```
