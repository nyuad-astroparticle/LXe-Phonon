# Bessel Function Estimation

We have found an estimate of the solution using bessel functions. Now it qould be a good idea to evaluate something for once!

```python
# Get some necessary modules
import numpy as np
import math
import scipy
import matplotlib.pyplot as plt
import scipy.constants as const
import scipy.integrate as integrate
from scipy.special import jv # Bessel function
from decimal import *

getcontext().prec = 28
# Define the Fundamental solution
def g(r,t,n=100,tau=1,lam=1):
    SUM     = Decimal(0)
    alpha   = Decimal(float(tau*lam*r))
    beta    = Decimal(float(tau*t))
    for i in range(0,n+1):
        SUM = SUM + beta**Decimal(i)/Decimal(math.factorial(int(i+1))) * Decimal(jv(2*i,2*float(alpha)))

    return Decimal.exp(-beta) * SUM * Decimal(tau) / Decimal(4 * np.pi**2 * r)

def G_(r,t,n=100,tau=1,lam=1):
    SUM     = 0
    alpha   = tau*lam*r
    beta    = tau*t
    for i in range(0,n+1):
        SUM = SUM + beta**i/math.factorial(int(i+1)) * jv(2*i,2*alpha)

    return np.exp(-beta) * SUM * tau / (4 * np.pi**2 * r)

G = np.vectorize(g)
```

```python
# Now let's plot the fundamental solution

# Create the plot elements
fig = plt.figure(figsize=(10,5),dpi=200)
ax  = fig.add_subplot(111)
ax.set_xlabel('Distance')
ax.set_ylabel('Pressure')

# Get the appropriate sizes
T = 10
R = 25
Npts = 1000
ZERO = 1e+1
t = np.linspace(ZERO,200,Npts)
r = np.linspace(ZERO,1000,Npts)

# Plot the actual thing
GG = G(R,t,n=130)
ax.plot(t,GG/max(GG))
# ax.set_xlim(1,max(r))
# ax.set_ylim(-2000,2000)

# BB = B(R,t,n=130)
# ax.plot(t,BB/max(BB))

g = lambda x,x0=0,s=1:np.exp(-(x-x0)**2/(2*s**2))
ax.plot(t,g(t,x0=R,s=5))
```

```python
# So now that we finally have the fundamental solution for this equation,
# we proceed in performing the convolution with the actual source function

# Get a property file and return a dictionary with the info
def get_properties(fname):
    labels  = np.loadtxt(fname,usecols=0,dtype=str,delimiter='(')
    units   = np.loadtxt(fname,usecols=0,dtype=str,delimiter=':')
    values  = np.loadtxt(fname,usecols=1)

    return dict(zip(labels,values))         # Create a dictionary with the properties of the liquid

# Load the text file with the properties of Xenon
prop_Xe = get_properties('./fluids/LXE.txt')
prop_mu = get_properties('./particles/muon.txt')

# Now we will define the appropriate constants in SI units
K       = prop_Xe['bulk_modulus']
c       = prop_Xe['sound_speed']
Cp      = prop_Xe['specific_heat_p']
beta    = prop_Xe['thermal_expansion']
mu      = prop_Xe['viscocity']
rho_0   = prop_Xe['rest_density']
I       = prop_Xe['ionization_potential']
Z       = prop_Xe['atomic_number']
M       = prop_Xe['molar_mass']
n       = rho_0/M*const.Avogadro                        # Particle density
s       = n**(-1/3)                                     # Inter-particle distance 
v       = prop_mu['speed']                              # Speed of particle
z       = prop_mu['charge']/const.elementary_charge     # Numer Charge of compound
cc      = const.c                                       # Speed of light
ne      = const.Avogadro * Z * rho_0/M                  # Electron Densiry
me      = const.electron_mass                           # Electron Mass
e       = -const.elementary_charge                      # Electron Charge
e0      = const.epsilon_0                               # Permitivity
bb      = v/cc                                          # Particle Relative speed

# dE/dx the bethe bloch formula for the particle
dEdx    = 4*np.pi/me * ne * z**2/v**2 *(e**2/(4*np.pi*e0))**2* (np.log(2*me*v**2/(I*(1-bb**2))) - bb**2)

# Control Constants
# They are scaling factors for the nondimensionalization
# To be altered each time so that to assist with the process
lam = 1e3
tau = 1e3

# Derived constants
w0      = K/mu                              # Attenuation Frequency
T       = tau/w0                            # Standard Time unit
L       = lam*T*c                           # Standard Length unit
P       = beta*v**2*L**3*T/(s**5*Cp)*dEdx   # Standard Pressure Unit

# Nondimensionalized source function
t0 = 0
f  = lambda r,z,t: (z-v*(t-t0)*T/L) * np.exp(-(L/s)**2/2 *(r**2+ (z-v*(t-t0)*T/L)**2))*1e4
```

## Integral

Now the integral we want to perform is the following:

$$u = \int_{\mathbb{R}^+} ds\ \int_{\mathbb{R}^3} d^3y\ G(\vec{x}-\vec{y},t-s)f(y,s)$$

In other words this is simply the convolution of the fundamental solution in the nondimensionalized regime with the source function define above. It is worth noting that we can do some simplification to the integral by recognizing that:

$$G(\vec{x},t) = G(|\vec{x}|,t)$$

That way our numerical integration puzzle is going to appear much easier!


```python
# Now let's plot the Bessel Functions

# Create the plot elements
fig = plt.figure(figsize=(10,5),dpi=200)
ax  = fig.add_subplot(111)
ax.set_xlabel('x')
ax.set_ylabel('y')

# Get the appropriate sizes
Npts = 1000
n = 2
ZERO = 1e-0
x = np.linspace(ZERO,100,Npts)

ja = lambda n,z: np.sqrt(2/(np.pi*z))*np.cos(z-np.pi/4*(2*n+1))

# Plot the actual thing
ax.plot(x,jv(n,x),ls='-', c='k',label='Bessel Function')
ax.plot(x,ja(n,x),ls='--',c='r',lw=1,label='Far Field Approximation')

ax.legend(frameon=False)
# ax.set_xlim(1,max(r))
# ax.set_ylim(-2000,2000)
```
