# Energy Spectrum of Hamiltonian

Here we create a plot of the enrgy spectrum of the hamiltonian of the quantized damped wave. It is important to know what in order to find where the spectrum is real we need to combine both phonon and antiphonon states. Check it out!

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# Model Parameters
Npts = 100
L = 1
k = np.linspace(0,3/L)

f1 = lambda k: 1j *L*k**2 + (k**2 - L**2* k**4 + 0j)**0.5
f2 = lambda k: 1j *L*k**2 - (k**2 - L**2* k**4 + 0j)**0.5

g1 = lambda k: -1j *L*k**2 + (k**2 - L**2* k**4 + 0j)**0.5
g2 = lambda k: -1j *L*k**2 - (k**2 - L**2* k**4 + 0j)**0.5


# Phonons
kk = np.concatenate((k,k),axis=0)
xx = np.concatenate((f1(k).real,f2(k).real),axis=0)
yy = np.concatenate((f1(k).imag,f2(k).imag),axis=0)

# The figure
fig = plt.figure(figsize=(10,8),dpi=70)
ax  = fig.add_subplot(111,projection='3d')

ax.plot(xx,yy,kk)
```
